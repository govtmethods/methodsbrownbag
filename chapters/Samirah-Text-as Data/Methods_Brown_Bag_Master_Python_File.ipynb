# Else and If Statements

If/Elif/Else Pattern


```python
# if <condition>:
#     <do something>
# elif <another condition>:
#     <do something else>
# else:
#     <fallback>
```

```python
x = 90

if x > 10:
    msg = "x is big"
elif x == 10:
    msg = "x is exactly 10"
else:
    msg = "x is small"

print(msg)
```

Common gotchas & rules

* End each if, elif, else line with a colon :
* Indentation defines the code block (use 4 spaces; be consistent)
* Conditions are booleans (True or False)
* Strings are case-sensitive ("Canada" ≠ "canada")

# Loops

For vs While

* Use **for** when you know how many iterations (items in a list/range).

* Use **while** when you loop until a condition changes (e.g., countdown, user input).

**While Loop**

```python
# while <condition>:
#     <do stuff>
#     <update something>  # prevents an infinite loop

countdown = 3
while countdown > 0:
    print("T-minus", countdown)
    countdown -= 1  # update!
print("Liftoff")
```

**For Loop **

Use when you know how many items/iterations.

* With numbers: range(start, stop, step) (stop is exclusive).

```python
# for with range(start, stop, step) — stop is exclusive.
for i in range(1, 6):  # 1,2,3,4,5
    print("i =", i)
```

```python
for char in "Georgetown":
  print(char + "!")
```

# Lists and Dictionaries

**Lists**

Lists use square brackets [] and have order. They can hold any mix of items and allow duplicates.
* Create/see length: use len(...) to count items.
* Access: zero-based indexing (items[0]) and slicing (items[a:b:c]; stop is exclusive).

Common actions:
* Add one item: .append(...)
* Add many items: .extend([...])
* Insert at a position: .insert(index, value)
* Remove & return last (or by index): .pop() / .pop(i)
* Membership test: x in items

When to use: ordered data, sequences (e.g., tokens in a sentence, a list of URLs to scrape).

**Dictionaries**

Dictionaries use curly braces {} and store key → value pairs. Keys are unique; order isn’t the point (though Python preserves insertion order).
Common Actions:
* Create: {key: value, ...}
* Add / update: dict[key] = value ← (this is how you “append” to a dict)
* Merge: .update({...})

Lookup Data:
* Direct: dict[key] (errors if missing) (Good for: cases where a missing key means your data is broken and you want to fail fast (e.g., every row must have a “year” field)
* Safe: dict.get(key, default) (If the key is missing → you get the default you provided (no error))

When to use: fast lookups, mappings (e.g., country → capital, word → count, url → last_seen).

Important! --> dictionaries are case sensitive (“canada” ≠ “Canada”)

**Slicing and Filtering**

```python
# Slicing & Filtering Lists (start:stop:step)

fruits = ["apple", "banana", "cherry", "date", "elderberry",
          "fig", "grape", "grapefruit", "strawberry", "peach", "raspberry", "watermelon", "lychee",
          "pear", "acai"]

# Filtering with a for-loop
short_fruit = []
long_fruit  = []

for fruit in fruits:
    if len(fruit) < 6:
        short_fruit.append(fruit)
    else:
        long_fruit.append(fruit)

print("Short-name fruits:", short_fruit)
print("Long-name fruits:", long_fruit)
```

```python
# Slicing examples
print("First 3:", fruits[:3])          # start defaults to 0
print("Last 3:", fruits[-3:])          # negative indices from the end
print("Every other:", fruits[::2])     # step = 2
print("Middle block (3..6):", fruits[3:7])  # stop is exclusive
print("Reversed:", fruits[::-1])       # step = -1
```

# APIs

**What:** An Application Programming Interface is a doorway for apps (and you) to request structured data from a service.

**Why it matters for Text-as-Data:** APIs give you clean JSON you can turn into DataFrames, skip messy scraping, and stay within provider rules.

**How:**

* You send an HTTP request to an endpoint (a URL).
* You include parameters that specify what you want.
* You get a response with a status code and a JSON body (usually).
* You read the JSON, extract fields, and (later) save to CSV.

**Security & keys (government/open data):**

* Some APIs need an API key. Treat it like a password.
* Do not paste keys into public Colabs or GitHub.
* Keep keys offline or in a private place


**JSON**

Javascript Object Notation files store structured data in a simple-ish and human readable way. In python, we usually use the json library (?) to interact with JSon files. Very popular for exchanging data with servers, storing metadata along with labels, etc.

# Web Scraping

```python
#Load All These Packages (every time you restart colab)
import requests
from bs4 import BeautifulSoup
import pandas as pd
from time import sleep
```

**The American Presidency Project**

Collection of speeches by American presidents and vice-presidents, hosted by the University of California, Santa Barbara. No APIs needed!

```python
#Load Request Page

base_url = 'https://www.presidency.ucsb.edu'
root_page =  '/documents/app-categories/spoken-addresses-and-remarks/presidential/eulogies?items_per_page=10'
headers = {'User-Agent': 'Mozilla/5.0'} #common practice to include 'Mozilla/5.0' as a value to indicate compatibility with Mozilla-based browsers
# send request, access webpage
response = requests.get(base_url + root_page, headers=headers)
# BeautifulSoup turns the raw html into a searchable tree - gets you the JSON
# response.text contains the page content as raw html
soup = BeautifulSoup(response.text, 'html.parser')

# Syntax: soup.find_all(tagname, attrs = {key:value, key:value})
```

```python
#Extract URLs of all articles
# use .find_all() method to find all instances that match the pattern
# under 'div' find all subcontent with class attribute = field-title
links = soup.find_all('div', attrs = {'class':'field-title'})

len(links) # checks how many links got extracted from that webpage
```

```python
links[0] # checking to see what the first scraped link was, use to see if it worked correctly
```

```python
# soup's .a.get('href) looks for tag <a> and extract href attribute
links[0].a.get('href')
```

```python
article_urls = [link.find('a').get('href') for link in links]
article_urls
```

```python
# Extract Articles

#try with 1 first to test code

article = requests.get(base_url + article_urls[0]) # getting specifically the first article
soup = BeautifulSoup(article.text, "html.parser")
content = soup.find('div', attrs = {'class':'field-docs-content'}).text # the stuff in the curly brackets comes from the html file
```

```python
# extract all articles and meta data
Speech =[]
for url in article_urls:
  article = requests.get(base_url + url)
  soup_doc = BeautifulSoup(article.text, "html.parser")
  title = soup_doc.find('div', attrs={'class':'field-ds-doc-title'}).text
  date = soup_doc.find('div', attrs = {'class':'field-docs-start-date-time'}).text
  content = soup_doc.find('div', attrs = {'class':'field-docs-content'}).text
  candidate = soup_doc.find('h3', attrs={'class':'diet-title'}).text
  Speech.append({'title': title, 'candidate': candidate,'content': content, 'date': date, 'url': url})
  sleep(0.5) #asking python to sleep for 0.5 seconds after scraping a text. if you dont, site might think you are a bot
```

```python
# Create a DataFrame from the list of dictionaries
df_speeches = pd.DataFrame(Speech)

df_speeches
```

```python
#storing this dataframe on google drive

data_path = '/content/drive/My Drive/'

# use .to_csv() method to save the data to drive
df_speeches.to_csv(data_path + 'speeches.csv', index=False)
```

```python
# run a for loop to scrape all article urls
links_all = []
for page in range(0, 6):  # Pages 0–5
    #using f-string inserting page value
    base_url = 'https://www.presidency.ucsb.edu'
    page_url = f"/documents/app-categories/spoken-addresses-and-remarks/presidential/eulogies?items_per_page=10&page={page}"
    response = requests.get(base_url + page_url)
    soup = BeautifulSoup(response.text, 'html.parser')
    links = soup.find_all('div', attrs = {'class':'field-title'})
    links_all.extend([link.find('a').get('href') for link in links])
    sleep(0.5)

    ##so this method only grabs us the links. not the speeches themselves!
```
